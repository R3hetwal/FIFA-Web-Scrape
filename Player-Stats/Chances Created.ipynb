```python
url = 'https://www.foxsports.com/soccer/fifa-world-cup-men/stats?category=control&sort=cc&season=2022&sortOrder=desc&groupId=12'
```

```python
import requests
```

```python
html = requests.get(url).text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tbody = soup.find('tbody')
```

```python
table = tbody.find_all('tr')
```

```python
table[0]
```

```python
tds = table[0].find_all('td')
```

```python
for item in range(0, len(table)):
    print(table[item])
```

```python
for item in range(0, len(table)):
    tds=(table[item]).find_all('td')
    print(tds)
```

```python
complete_data = []
for item in range(0, len(table)):
    player_data = []
    table_row = table[item]
    tds = table_row.find_all('td') # one table row has multiple table data
    for item in tds:
        player_data.append(item.text) # extract data of each player
   
    complete_data.append(player_data) # append all data of each  player on a list
```

```python
clean_data = list(map(lambda x: x[1:20], complete_data))
```

```python
clean_data
```

```python
# !pip install pandas
# !pip install matplotlib
# !pip install seaborn
```

```python
import pandas as pd # as pd is alias name convention
```

```python
df = pd.DataFrame(clean_data)
```

```python
df.columns = col_names
```

```python
df = df.replace('\n', '', regex=True)
```
